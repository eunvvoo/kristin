```python
import os
import shutil
import json
```

```python
path = '/home/compu/Documents/exports/json/'
file_list = os.listdir(path)    
data = []

# json 파일 병합
for file in file_list:
    with open(path + file, 'r') as json_file:
        data.append(json.load(json_file))
        
# json 파일 저장
with open("/home/compu/Documents/exports/newfile.json", "w") as new_file:
    json.dump(data, new_file)
```

```python
#slip_on을 fashion_sneakers에 병합
key_errors = []

for d in data:  
    try:
        if d['shoes_type'] == 'slip_on':
            print(d['shoes_type'] + "->")
            d['shoes_type'] = 'fashion_sneakers'
            print(d['shoes_type'])
    except KeyError:
        key_errors.append(d['id'])
```

```python
path = '/home/compu/Documents/exports/'
key_errors = []
class_list = ['fashion_sneakers', 'others']

for attr in class_list:
    if not os.path.exists(path+"item_merge/train/"+attr+"/"):
        os.makedirs(path+"item_merge/train/"+attr+"/")

        
for d in data:  
    try:
        if d["shoes_type"] in class_list:
            shutil.copyfile(path+'img/'+d["img_name"]+'.jpg', path+'item_merge/train/'+d["shoes_type"]+'/'+d["img_name"]+'.jpg')
        
    except KeyError:
        key_errors.append(d["id"])
    except Exception as e:
        print(e, d["id"])
```

```python
# 사진 랜덤하게 20% test로 옮기기
import random
dirs = ['fashion_sneakers/', 'others/']
path = '/home/compu/Documents/exports/item_merge/train/'
path2 = '/home/compu/Documents/exports/item_merge/test/'

for direct in dirs:
    if not os.path.exists(path2+direct):
        os.makedirs(path2+direct)
    file_list = os.listdir(path+direct)
    
    # 20% 나누기
    x = len(file_list) // 5
    test_dataset = random.sample(file_list, x)
    for test_image in test_dataset:
        shutil.move(path+direct+test_image, path2+direct+test_image)
    print(f"{direct} 총 {len(os.listdir(path+direct))}장 중 {x}장 이동")
```
