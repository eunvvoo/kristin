```python
path = '/home/compu/Documents/exports/'
key_errors = []
class_list = ["leather", "mesh/knit", "suede", "nylon"]

for attr in class_list:
    if (attr == "mesh/knit") & (not os.path.exists(path+"material_merge/train/mesh_knit")):
        os.makedirs(path+"material_merge/train/mesh_knit/")

    elif not os.path.exists(path+"material_merge/train/"+attr+"/"):
        os.makedirs(path+"material_merge/train/"+attr+"/")

        
for d in data:  
    try:
        for attr in d["material"]:
            if attr == "mesh/knit":
                shutil.copyfile(path+'img/'+d["img_name"]+'.jpg', path+'material_merge/train/'+'mesh_knit/'+d["img_name"]+'.jpg')

            elif attr in class_list:
                shutil.copyfile(path+'img/'+d["img_name"]+'.jpg', path+'material_merge/train/'+attr+'/'+d["img_name"]+'.jpg')
        
    except KeyError:
        key_errors.append(d["id"])
    except Exception as e:
        print(e, d["id"])
```

```python
# 사진 랜덤하게 20% test로 옮기기
import random
dirs = ["leather/", "mesh_knit/", "suede/","nylon/"]
path = '/home/compu/Documents/exports/material_merge/train/'
path2 = '/home/compu/Documents/exports/material_merge/test/'

for direct in dirs:
    if not os.path.exists(path2+direct):
        os.makedirs(path2+direct)
    file_list = os.listdir(path+direct)
    
    # 20% 나누기
    x = len(file_list) // 5
    test_dataset = random.sample(file_list, x)
    for test_image in test_dataset:
        shutil.move(path+direct+test_image, path2+direct+test_image)
    print(f"{direct} 총 {len(os.listdir(path+direct))}장 중 {x}장 이동")
```
