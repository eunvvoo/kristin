```python
import os
import shutil
import json
```

```python
path = '/home/compu/Documents/exports/json/'
file_list = os.listdir(path)    
data = []

# json 파일 병합
for file in file_list:
    with open(path + file, 'r') as json_file:
        data.append(json.load(json_file))
        
# json 파일 저장
with open("/home/compu/Documents/exports/newfile.json", "w") as new_file:
    json.dump(data, new_file)
```

```python
# 병합한 json 파일 로드
with open('/home/compu/Documents/exports/newfile.json', 'r') as json_file:
        data = json.load(json_file)
```

```python
#shark, combat_boots를 tread에 병합
key_errors = []

for d in data:  
    try:
        d['sole'] = [attr.replace('shark','tread') for attr in d['sole']]
        d['sole'] = [attr.replace('combat_boots','tread') for attr in d['sole']]
        print(d['sole'])
    except KeyError:
        key_errors.append(d['id'])
```

```python
# json 파일 저장
with open("/home/compu/Documents/exports/newfile_sole_merge.json", "w") as new_file:
    json.dump(data, new_file)
```

```python
path = '/home/compu/Documents/exports/'
key_errors = []
class_list = ["sport", "ugly", "tread", "cup_sole"] #flat 제외

for attr in class_list:
    if not os.path.exists(path+"sole_merge/train/"+attr+"/"):
        os.makedirs(path+"sole_merge/train/"+attr+"/")

        
for d in data:  
    try:
        for attr in d["sole"]:
            if attr in class_list:
                shutil.copyfile(path+'img/'+d["img_name"]+'.jpg', path+'sole_merge/train/'+attr+'/'+d["img_name"]+'.jpg')
        
    except KeyError:
        key_errors.append(d["id"])
    except Exception as e:
        print(e, d["id"])
```

```python
# 사진 랜덤하게 20% test로 옮기기
import random
dirs = ["sport/", "ugly/", "tread/", "cup_sole/"]
path = '/home/compu/Documents/exports/sole_merge/train/'
path2 = '/home/compu/Documents/exports/sole_merge/test/'

for direct in dirs:
    if not os.path.exists(path2+direct):
        os.makedirs(path2+direct)
    file_list = os.listdir(path+direct)
    
    # 20% 나누기
    x = len(file_list) // 5
    test_dataset = random.sample(file_list, x)
    for test_image in test_dataset:
        shutil.move(path+direct+test_image, path2+direct+test_image)
    print(f"{direct} 총 {len(os.listdir(path+direct))}장 중 {x}장 이동")
```

```python
#폴더 다시 합치기
dirs = ["sport/", "ugly/", "tread/", "cup_sole/"]
path = '/home/compu/Documents/exports/sole_merge/train/'
path2 = '/home/compu/Documents/exports/sole_merge/test/'

for direct in dirs:
    train_dataset = os.listdir(path+direct)
    for train_image in train_dataset:
        shutil.move(path+direct+train_image, path+train_image)
    
    test_dataset = os.listdir(path2+direct) 
    for test_image in test_dataset:
        shutil.move(path2+direct+test_image, path2+test_image)

for direct in dirs: 
    shutil.rmtree(path+direct)
    shutil.rmtree(path2+direct)
```
